```python
import seaborn as sns
import matplotlib.pyplot as plt

# Memilih fitur numerik
numerical_features = ['carat', 'depth', 'table', 'price', 'x', 'y', 'z']

# Membuat boxplot untuk setiap fitur numerik
for feature in numerical_features:
    plt.figure(figsize=(8, 6))
    sns.set(style="whitegrid")  # Mengatur gaya plot

    # Membuat boxplot dengan warna pastel
    sns.boxplot(x=df[feature], color="lightblue", width=0.5)

    plt.title(f'Boxplot of {feature}', fontsize=16)
    plt.xlabel(f'{feature}', fontsize=14)
    plt.show()
```

#Scatter Plot

Prompt ChatGPT : "beri code untuk menampilkan scatter plot untuk semua fitur numerik!"

```python
# Memilih fitur numerik
numerical_features = ['carat', 'depth', 'table', 'price', 'x', 'y', 'z']

# Membuat scatter plot untuk setiap pasangan fitur numerik
sns.set(style="ticks")  # Mengatur gaya plot
sns.pairplot(df[numerical_features])
plt.show
```

#Heatmap

Prompt ChatGPT : "beri code untuk menampilkan heatmap dari fitur2 numerik!"

```python
# Memilih fitur numerik
numerical_features = ['carat', 'depth', 'table', 'price', 'x', 'y', 'z']

# Menghitung korelasi antara fitur numerik
correlation_matrix = df[numerical_features].corr()

# Membuat heatmap dari korelasi
plt.figure(figsize=(10, 8))
sns.set(font_scale=1.2)
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Heatmap of Numerical Features Correlation', fontsize=16)
plt.show()
```
