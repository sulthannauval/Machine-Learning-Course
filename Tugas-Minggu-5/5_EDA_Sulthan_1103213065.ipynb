Prompt ChatGPT : "Beri kode untuk mengisi baris dari fitur "Kurzarbeit hours per week" yang memiliki null value dengan nilai "0"."

Prompt ChatGPT : "Beri kode untuk menampilkan pie diagram chart untk fitur Age dengan min 20, max 69 dengan range 10 per umur"

```python
import matplotlib.pyplot as plt

# Tentukan rentang umur yang diinginkan
umur_min = 20
umur_max = 69
range_umur = 10

# Buat bins untuk rentang umur
bins_umur = list(range(umur_min, umur_max + range_umur, range_umur))

# Gunakan cut untuk membagi data ke dalam bins umur
df['Age Range'] = pd.cut(df['Age'], bins=bins_umur, right=False)

# Hitung jumlah pengamatan dalam setiap bins umur
count_per_range = df['Age Range'].value_counts().sort_index()

# Buat pie chart
plt.figure(figsize=(8, 6))
count_per_range.plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=plt.cm.tab10.colors)
plt.title('Distribusi Umur')
plt.ylabel('')
plt.axis('equal')  # Agar lingkaran terlihat lingkaran
plt.show()
```

Prompt ChatGPT : "Beri kode untuk menampilkan baris data yang memiliki Age '51-60', '60+'."

```python
# Filter baris data yang memiliki umur antara '51-60' dan '60+'
baris_umur_51_60 = df[df['Age'].between(51, 60)]
baris_umur_60_plus = df[df['Age'] >= 60]

# Gabungkan kedua hasil filter menggunakan metode concat
baris_umur = pd.concat([baris_umur_51_60, baris_umur_60_plus])

# Tampilkan baris data yang memiliki umur '51-60' dan '60+'
baris_umur
```

Prompt ChatGPT : "beri code untuk analisis korelasi dan analisis ouitlier"

```python
import seaborn as sns

# Analisis Korelasi
korelasi = df.corr()

# Visualisasi Matriks Korelasi dengan heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(korelasi, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Matriks Korelasi antar Fitur')
plt.show()
```

```python
# Analisis Outlier
fitur_numerik = df.select_dtypes(include=['int64', 'float64']).columns

# Visualisasi boxplot untuk masing-masing fitur numerik
plt.figure(figsize=(16, 12))
for i, col in enumerate(fitur_numerik, 1):
    plt.subplot(3, 3, i)
    sns.boxplot(x=df[col])
    plt.title(f'Boxplot dari {col}')

plt.tight_layout()
plt.show()
```
